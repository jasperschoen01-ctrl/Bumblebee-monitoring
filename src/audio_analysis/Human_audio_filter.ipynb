```python
import torch

# 1. Load the Offline Neural Network
# (It will download a ~2MB file the first time, then run locally forever)
model, utils = torch.hub.load(repo_or_dir='snakers4/silero-vad',
                              model='silero_vad',
                              force_reload=True,
                              trust_repo=True)

# Unpack the helper functions that come with the model
(get_speech_timestamps, save_audio, read_audio, VADIterator, collect_chunks) = utils

# 2. Load your 6-second audio file
# PyTorch reads it and forces it to the 16,000 Hz that the AI requires
file_path = 'data/original/20260324_123001.WAV'
wav = read_audio(file_path, sampling_rate=16000)

# 3. Scan the file for human voices
# return_seconds=True makes it output nice, human-readable timestamps
speech_timestamps = get_speech_timestamps(wav, model, sampling_rate=16000, return_seconds=True)

# 4. Print the results
print("AI Voice Detection Complete!")
if not speech_timestamps:
    print("No human voices detected.")
else:
    print("Human voice detected at the following times:")
    for speech_event in speech_timestamps:
        print(f"Start: {speech_event['start']} sec  --->  End: {speech_event['end']} sec")
```

```python
import soundfile as sf
import numpy as np
import os

# 1. Define the path to your Demucs folder (update this to your actual path)
stem_folder = "separated/htdemucs/20260324_125100/"
output_folder = "separated"

# 2. Load the stems you want to keep (leaving 'vocals.wav' behind)
other_audio, sr = sf.read(stem_folder + "other.wav")
bass_audio, _ = sf.read(stem_folder + "bass.wav")
drums_audio, _ = sf.read(stem_folder + "drums.wav")

# 3. Mathematically mix them together by adding the arrays
# (This perfectly reconstructs the audio, minus the singer)
vocal_free_mix = other_audio + bass_audio + drums_audio

# 4. Save the new file to your computer
output_filename = "radio_without_vocals_20260324_125100.wav"
output_path = os.path.join(output_folder, output_filename)
sf.write(output_path, vocal_free_mix, sr)

print(f"Success! Saved new file: {output_filename}")
```
